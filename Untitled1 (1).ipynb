```python
%matplotlib inline
import imageio
import requests
import matplotlib.pyplot as plt
import IPython.display as dp
```

```python
from PIL import Image
import IPython.display as dp

img_path = '/content/image.webp'  # Local file path
image = Image.open(img_path)  # Open the image
dp.display(image)  # Display the image
```

```python
source_img = imageio.imread(img)
```

```python
import numpy as np
def grayscaleing(rgb):
  return np.dot(rgb[...,:3],[0.299,0.587,0.114])
gryscl_img = grayscaleing(source_img)
```

```python
inv_img = (255 - gryscl_img)
plt.imshow(inv_img)
```

```python
import scipy.ndimage
blurred_img = scipy.ndimage.filters.gaussian_filter(inv_img, sigma=5)
plt.imshow(blurred_img)
```

```python
def dodging(blur_img,gryscl_img):
  resultant_dodge=blur_img*255/(255-gryscl_img)
  resultant_dodge[resultant_dodge>255]=255
  resultant_dodge[gryscl_img==255]=255
  return resultant_dodge.astype('uint8')
```

```python
target_img= dodging(blurred_img, gryscl_img)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.imshow(target_img, cmap='gray')
```

```python
import cv2
import numpy as np
import gradio as gr
from PIL import Image

def convert_to_sketch(image):
    # Convert PIL image to OpenCV format
    image = np.array(image)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Invert grayscale
    inverted = 255 - gray

    # Apply Gaussian Blur
    blurred = cv2.GaussianBlur(inverted, (21, 21), 0)

    # Invert the blurred image
    inverted_blur = 255 - blurred

    # Create the sketch
    sketch = cv2.divide(gray, inverted_blur, scale=256.0)

    # Convert back to PIL format
    return Image.fromarray(sketch)

# Create Gradio Interface
iface = gr.Interface(
    fn=convert_to_sketch,
    inputs=gr.Image(type="pil"),
    outputs="image",
    title="Sketch Converter",
    description="Upload an image and get a sketch version instantly!"
)

# Launch the app
iface.launch()
```
